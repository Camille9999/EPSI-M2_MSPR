```python
import os
from dotenv import load_dotenv
load_dotenv()
PATH_DATA = os.getenv('PATH_DATA')
```

```python
import pandas as pd

pd.options.display.max_columns = 0

def show_df(df):
    display(pd.concat([df.head(2), df.tail(1)]))
    print(df.shape)
```

```python
year = 2012
data_path = f"{PATH_DATA}/source/{year}_eCO2mix_RTE_Annuel-Definitif_{year}.xls"

df = pd.read_csv(data_path, sep="\t", skipfooter=2, encoding="latin1")

show_df(df)
```
